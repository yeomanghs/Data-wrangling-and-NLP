```python
import json
```

```python
#spacy
jsonFile = "D:\\Users\\figohjs\\Documents\\NLP\\NER\\data\\training\\2021-01-26_SpaCyPrediction200.json"

munirahList = []

with open(jsonFile) as input_file:
    for row in input_file.readlines():
        munirahList.append(json.loads(row))   

textList = [i['Description'] for i in munirahList]
tagList = [i['RealTag'] for i in munirahList]
#extract munirah prediction
predMunirahList = [i['PredTagMunirah'] for i in munirahList]


#crf
jsonFile = "D:\\Users\\figohjs\\Documents\\NLP\\NER\\data\\training\\2021-01-26_CRFv1Prediction200.json"

crfList = []

with open(jsonFile) as input_file:
    for row in input_file.readlines():
        crfList.append(json.loads(row))   

#extract munirah prediction
predCRFList = [i['PredTagCRF'] for i in crfList]
```

```python
def printText(no):    
#     print(textList[no] + '\n')
    return textList[no]

def printPrediction(no):
    #get org or person only
    tagList2 = [[i[0], i[1]] for i in tagList[no] if i[1] in ['ORG', 'PERSON']]
    print("Real Tag:\n%s\n"%tagList2)
    print("SpaCy Prediction:\n%s\n"%predMunirahList[no])
    print("CRF Prediction: \n%s\n"%predCRFList[no])
```

### CRF has higher precision than SpaCy - give less false positives

```python
printText(1)
```

```python
printPrediction(1)
```
